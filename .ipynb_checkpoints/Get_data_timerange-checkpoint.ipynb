# Data Gathering

### Specify time range

```python
##### Specify time range #####
startDate = datetime(2018, 11, 1, 0, 0)
endDate = datetime(2018, 11, 2, 0, 0)
startTime = int(startDate.timestamp())*1000;  print(startTime)
endTime = int(endDate.timestamp())*1000;  print(endTime)
```

## Gather data for the weather station

```python
for i in range(0,len(metricsList)):
    # Select metric
    metricID = metricsList[i]

    # Set begin
    begin = startTime;

    # For time measurement purposes
    t = time.time()

    first = True;
    while(begin < endTime):
        # Set end to be an hour later [ms]
        end = begin + 60*60*1000
        # Create URL
        myURL = 'https://idlab-iot.tengu.io/api/v1/scopes/{scope}/things/{thing}/metrics/{metric}/events?from={start}&to={stop}'\
                    .format(scope = scopeID, thing = thingID, metric = metricID, start = begin, stop = end)
    #     print(myURL)
        # Update begin timestamp for next loop
        begin = end;

        # Execute query
        response = requests.get(myURL, headers=headers);
        data = response.json() 
        column_names = data['columns']; #print(column_names)
        values = data['values'] 

        # First loop: make the table
        if first:
            df = pd.DataFrame.from_dict(values)
            df.columns = column_names
            first = False;
        # All other loops, append
        else:
            df_extra = pd.DataFrame.from_dict(values)
            df_extra.columns = column_names
            df = df.append(df_extra,ignore_index=True)

    elapsed = time.time() - t
    print(str(elapsed) + " sec")


    # Select timestamp and value from results
    weather_light_data = df[['time','value']]
    # Convert object to float (BatteryVoltHR is an object for example)
    weather_light_data["value"] = pd.to_numeric(weather_light_data.value, errors='coerce')
    # Rename the column
    weather_light_data.rename(columns={'value':metricsList[0]}, inplace=True)
    # Set timestamp as index
    weather_light_data = weather_light_data.set_index(pd.DatetimeIndex(pd.to_datetime(weather_light_data.time, unit='ms')))

    # save to CSV
    weather_light_data.to_csv('Light_181122_181128.csv')

    # Print values for visual checking
    weather_light_data
```
