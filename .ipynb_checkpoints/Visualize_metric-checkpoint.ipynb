# Format the data
Once the data is received, it's in a dictionary type. In order to be able to work with it properly, I want to convert it to a pandas DataFrame.
First the column names and the values are extracted from the dictionary. Then the values are converted to a DataFrame and the column names are used to rename the columns in the DataFrame.
I also convert the timestamp in milliseconds to a more readable timestamp in the DateTime format. The time-column is then converted to be used as an index for the table.

```python
data = response.json() # This is a dictionary type
column_names = data['columns'] #This is a list
values = data['values'] #This is a list
print(values)
```

```python
df = pd.DataFrame.from_dict(values)
df.columns = column_names
# df['time']= pd.to_datetime(df.time, unit='ms')
df.time = pd.to_datetime(df.time, unit='ms')
df.head()
# df['time'][0]
```

```python
df = df.set_index(pd.DatetimeIndex(df['time']))
df.info()
df.head()
```
