### Get available metrics for certain thing

```python
thingID = 'lora.3432333864378618'

myUrlMetrics = 'https://idlab-iot.tengu.io/api/v1/scopes/{scope}/things/{thing}/metrics'.format(scope = scopeID, thing = thingID)

headers = {
    'accept': 'application/json',
    'authorization': 'Bearer {0}'.format(myToken),
}

response = requests.get(myUrlMetrics, headers=headers)
response.content
data = json.dumps(response.json()) # convert list to string
metrics = pd.read_json(data)
print(metrics)
```

### Get available data for certain thing

```python
metricsList = metrics['id'].tolist()
```

```python
# len(metrics)
```

```python
startDate = datetime.datetime(2018, 11, 17, 0, 0)
endDate = datetime.datetime(2018, 11, 18, 0, 0)
startTime = int(startDate.timestamp())*1000;  print(startTime)
endTime = int(endDate.timestamp())*1000;  print(endTime)

metricID = metricsList[0]
myURL = 'https://idlab-iot.tengu.io/api/v1/scopes/{scope}/things/{thing}/metrics/{metric}/events?from={start}&to={end}'\
                .format(scope = scopeID, thing = thingID, metric = metricID, start = startTime, end = endTime)
print(myURL)
response = requests.get(myURL, headers=headers); 
print(type(response))
# print(response.content);
assert response.content != 'unauthorized'

data = response.json() # This is a dictionary type
print(type(data))
column_names = data['columns']; print(column_names) #This is a list 
values = data['values'] #This is a list
df = pd.DataFrame.from_dict(values)
df.columns = column_names
df
```

```python
all_data = df[['time','value']]
all_data.rename(columns={'value':metricsList[0]}, inplace=True)

print(len(all_data))
all_data.head()
df = all_data
```

```python
df.time = pd.to_datetime(df.time, unit='ms')
df = all_data.set_index(pd.DatetimeIndex(df['time']))
df.info()
df.head()
```

```python
df['airquality.no2'].plot(title='airquality.no2')
plt.ylabel('airquality.no2')
plt.show()
```
